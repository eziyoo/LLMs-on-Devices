```python

# Step 1: Load and save the dataset
from datasets import load_dataset
```

```python
# Load wikitext2 dataset
dataset = load_dataset("wikitext", "wikitext-2-raw-v1")
```

```python
# Save to CSV
dataset["train"].to_csv("Dataset/wikitext2 dataset/wikitext2_train.csv", index=False)
dataset["test"].to_csv("Dataset/wikitext2 dataset/wikitext2_test.csv", index=False)
```
